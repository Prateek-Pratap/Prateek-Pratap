## 3. ***Data Wrangling***

### Data Wrangling Code

After analyzing the dataset for null values, we found that four columns contain null values: 'children', 'country', 'agent', and 'company'. The respective counts of null values for these columns are as follows: 'company' - 112,593, 'agent' - 16,340, 'country' - 488, and 'children' - 4. To handle these null values, we will fill the 'children' column with 0 and the 'country' column with 'other'. Since the 'agent' column has null values that account for 13.69% (16,340) of the data, we can tolerate them and replace them with 0. However, due to the high percentage of null values in the 'company' column (approximately 95% or 112,593), we will drop that column as it won't provide useful information.

After analyzing the dataset, it was found that there are 32,001 duplicate rows. To ensure the accuracy of our results, we have decided not to remove these duplicates. The reason for this is that when the duplicates were removed earlier, there were discrepancies in the counts and percentages of variables.

For example, when duplicate values were dropped for the year 2015, City Hotel had fewer bookings compared to Resort Hotel. However, without removing duplicates, the situation reversed, and City Hotel showed more bookings than Resort Hotel.

The same inconsistency occurred with customer types. After removing duplicates, the count and percentage for Transient were 71,982 and 82.369635, respectively, while for Transient-Party, the count and percentage were 11,724 and 13.415876, respectively. However, without removing duplicates, the counts and percentages changed to 89,613 and 75.059050 for Transient, and 25,124 and 21.043638 for Transient-Party, respectively.

````python
# Write your code to make your dataset analysis ready.

# Fill null values in the 'children' column with 0
df['children'].fillna(0, inplace=True)

# Fill null values in the 'country' column with 'other'
df['country'].fillna('other', inplace=True)

# Replace null values in the 'agent' column with 0
df['agent'].fillna(0, inplace=True)

# Drop the 'company' column due to a high percentage of null values
df.drop('company', axis=1, inplace=True)

# We will not remove the duplicate values(For reason See above)


#Code to remove the duplicates and just keep the first copy of the data if required.
#df.drop_duplicates(keep = 'first', inplace = True, ignore_index = True)'''

````

## **Outlier Analysis : ADR vs. Stay Visualization**

````python
# Chart - 12 visualization code

df_stay_adr = pd.DataFrame(df['adr'])
df_stay_length = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
df_stay_adr['days'] = df_stay_length

plt.figure(figsize=(12, 7))
plt.subplot(1, 2, 1)
plt.title('Outlier for ADR')
plt.boxplot(df_stay_adr['adr'], meanline=True)

plt.subplot(1, 2, 2)
plt.title('Outlier for Number of days a customer stayed')
plt.boxplot(df_stay_adr['days'], meanline=True)

plt.show()

# Removing the outlier or the rows where the ADR value is 0, since the ADR cannot be 0.
df_stay_adr.drop(df_stay_adr[(df_stay_adr['adr'] > 5000) | (df_stay_adr['adr'] <= 0)].index, inplace=True, axis=0)

# Plotting a scatter plot that shows the relation between the ADR and the days the customer had stayed
plt.figure(figsize=(28, 3))
plt.scatter(x=df_stay_adr[['days']], y=df_stay_adr['adr'], marker='*', alpha=0.5)
plt.title('ADR vs stay')
plt.xlabel('Number of days stayed')
plt.ylabel('ADR')
plt.show()
````

The insights that can be derived from the above charts are:

1. Outlier for ADR: The box plot helps identify any extreme values or outliers in the Average Daily Rate (ADR). This information can be useful in understanding unusual pricing patterns or anomalies in the data.
